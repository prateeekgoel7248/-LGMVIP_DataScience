# Iris Flowers Classification ML Project :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris.tail()
```

```python
x = iris[['SepalLengthCm',"SepalWidthCm",'PetalLengthCm','PetalWidthCm']]
y = iris[['Species']]
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
y_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
regressor = LogisticRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#Now we will checking the testing data with the predictions
ml = pd.DataFrame({'Real': y_test['Species'], 'Predicted': y_pred})
ml
```

```python
classification_report(y_test.Species, y_pred)
```

```python
print('Accuracy of Model: ',accuracy_score(y_pred,y_test.Species)*100)
```
